### Q2. 2011년 부터의 연도별, 국가별 평균 기대수명을 구해주세요. (2011년도 포함되게 구합니다.)

* [필수 조건] groupby 나 pivot_table을 활용합니다. 
* groupby 로 구할 때 unstack()이라는 기능을 사용하여 아래와 같이 컬럼에 인덱스 값을 올려서 표기할 수 있습니다.
* 만약 행에는 '연도'가, 열에 '국가'가 들어있고 수치 데이터의 결과값이 아래와 같다면, 출력형태는 조금 달라도 괜찮습니다.

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
```

```python
# 2011년 행 찾기
df.loc[df['Year'] == 2011]
```

```python
# 행의 인텍스 214~273까지 연도별 국가별 평균 기대수명, groupby 사용
g_df1 = df.loc[214:].groupby(['Year','Country'])['Life_Expectancy'].mean().unstack()
g_df1
```

```python
g_df1.columns.values
```

```python
# 열 재조정
g_df2 = g_df1.copy()
g_df2.columns = g_df2.columns.values
g_df2 = g_df2.reset_index(level=0)
g_df2
```

```python
# 행의 인텍스 214~273까지 연도별 국가별 평균 기대수명, pivot_table 사용
pv_df1 = pd.pivot_table(df.iloc[214:],                # 피벗할 데이터프레임
                     index = 'Year',    # 행 위치에 들어갈 열
                     columns = 'Country',    # 열 위치에 들어갈 열
                     values = 'Life_Expectancy',     # 데이터로 사용할 열
                     aggfunc = 'mean')   # 데이터 집계함수
pv_df1
```

```python
pv_df1.columns.values
```

```python
# 열 재조정
pv_df2 = pv_df1.copy()
pv_df2.columns = pv_df2.columns.values
pv_df2 = pv_df2.reset_index(level=0)
pv_df2
```
